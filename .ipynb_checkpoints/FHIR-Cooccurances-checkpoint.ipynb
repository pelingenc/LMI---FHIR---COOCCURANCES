```python
import os
import re
import pandas as pd
import tkinter as tk
from tkinter import ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
import networkx as nx
from collections import defaultdict
import math



directory = 'C:/Users/begon/synthea/output/deneme1'

pattern = re.compile(r'(\"resourceType\": \"(Patient|Encounter|Condition|Observation|Procedure|Medication)\")')

classes = ['Patient', 'Encounter', 'Condition', 'Observation', 'Procedure', 'Medication']

current_resource_type = None
new_resource_type = None
resource_type = None
current_id = None
new_id = None
current_references = []
current_codes = []
current_patient_id = None

data = {'ResourceType': [], 'ID': [], 'References': [], 'Codes': [], 'PatientID': []}

counts = {resource_type: 1 for resource_type in classes}

INIT = False

for filename in os.listdir(directory):
    if filename.endswith('.json'):
        with open(os.path.join(directory, filename), 'r', encoding='utf-8') as f:
            for line in f:
                match = pattern.search(line)
                if '"resourceType": "' in line and not match:
                    resource_type = line.split('"')[3]
                    INIT = False
                if match:
                    if match.group(2) in classes:
                        new_resource_type = match.group(2)
                        INIT = True
                        next_line = next(f, None)
                        if next_line and '"id": "' in next_line:
                            new_id = next_line.split('"id": "')[1].split('"')[0]
                if current_resource_type == new_resource_type and '"reference": "urn:uuid:' in line and INIT:
                    reference = line.split(':')[3].split('"')[0]
                    current_references.append(reference)      
                if current_resource_type == new_resource_type and '"code": "' in line and INIT:     
                    code = line.split('"code": "')[1].split('"')[0]
                    current_codes.append(code)
                if current_resource_type != new_resource_type  and INIT:
                    if current_resource_type and current_id:
                        data['ResourceType'].append(current_resource_type)
                        data['ID'].append(current_id)
                        if current_resource_type == 'Patient':
                            current_patient_id = current_id
                        data['References'].append(tuple(current_references))
                        data['Codes'].append(tuple(current_codes))
                        data['PatientID'].append(current_patient_id)
                        current_references = []
                        current_codes = []
                    current_resource_type = new_resource_type    
                    current_id = new_id

df = pd.DataFrame(data)
grouped = df.groupby('PatientID')
aggregated_data = grouped.sum()
aggregated_data.to_csv('C:/Users/begon/Desktop/LMI PROJECT/Patient1.csv')

for patient_id, group in grouped:
    encounter_references = []
    for index, row in group.iterrows():
        if row['ResourceType'] == 'Encounter':
            patient = row['PatientID']
            encounter_id = row['ID']
            references = [encounter_id]
            codes = [code for code in row['Codes'] if re.match(r'^\d+$|^i$', code)]
            matching_rows = group[group['References'].apply(lambda x: encounter_id in x)]
            matching_codes = [code for codes in matching_rows['Codes'] for code in codes if re.match(r'^\d+$|^i$', code)]
            merged_codes = ','.join(set(codes + matching_codes))
            encounter_references.append({'Patient': patient, 'EncounterID': encounter_id, 'References': references, 'MergedCodes': merged_codes})
    encounter_references_df = pd.DataFrame(encounter_references)
    encounter_references_df.to_csv(f'C:/Users/begon/Desktop/LMI PROJECT/Encounter1/{patient_id}.csv', index=False)

directory = 'C:/Users/begon/Desktop/LMI PROJECT/Encounter1'
dfs = []
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        csv_path = os.path.join(directory, filename)
        df = pd.read_csv(csv_path)
        selected_columns = df.iloc[:, [1, -1]]
        dfs.append(selected_columns)

combined_df = pd.concat(dfs, ignore_index=True) 
combined_df = combined_df[combined_df['MergedCodes'] != '[]']
combined_df.reset_index(drop=True, inplace=True)
combined_df['MergedCodes'] = combined_df['MergedCodes'].apply(lambda x: x.strip('[]').split(',') if isinstance(x, str) else [])
combined_csv_path = 'Encounter_combined_data1.csv'
combined_df.to_csv(combined_csv_path, index=False)




def read_groups_from_csv(filename):
    groups = []
    data = pd.read_csv(filename)
    groups = [eval(code) for code in data['MergedCodes']]
    return groups

combined_csv_path = 'Encounter_combined_data1.csv'
groups = read_groups_from_csv(combined_csv_path)
all_codes = [code for group in groups for code in group]
unique_codes = set(all_codes)
options = [code.strip("[]' ") for code in unique_codes]
print(options)


def generate_graph(canvas, selected_groups, num_nodes_to_visualize):
    plt.clf()
    G = nx.Graph()
    for group in selected_groups:
        add_edges_within_group(G, group)
    node_counts = defaultdict(int)
    for group in selected_groups:
        for node in group:
            node_counts[node] += 1
    sorted_nodes = sorted(node_counts, key=node_counts.get, reverse=True)
    selected_nodes = sorted_nodes[:num_nodes_to_visualize]
    subgraph = G.subgraph(selected_nodes)

    if len(subgraph) < 10:
        pos = nx.spring_layout(subgraph)
    else:
        pos = nx.circular_layout(subgraph)

    node_size = [subgraph.degree(n)**2*10000/num_nodes_to_visualize**2 for n in subgraph.nodes()]
    if num_nodes_to_visualize > 0:
        nx.draw_networkx_nodes(subgraph, pos, node_size=node_size, node_color="w", edgecolors="k", linewidths=2.0 / num_nodes_to_visualize)
    else:
        nx.draw_networkx_nodes(subgraph, pos, node_size=node_size, node_color="w", edgecolors="k", linewidths=2.0)
    nx.draw_networkx_labels(subgraph, pos, labels={node: str(node) for node in subgraph.nodes()}, font_size=14)

    weights = [len(list(nx.common_neighbors(subgraph, u, v))) for u, v in subgraph.edges()]
    nx.draw_networkx_edges(subgraph, pos, width=[w**4 /num_nodes_to_visualize**2 for w in weights], edge_color="black")

    plt.axis("off")
    canvas.draw()

def on_dropdown_select(event):
    global selected_groups
    selected_code = dropdown.get()
    selected_groups = [group for group in groups if selected_code in group]
    num_nodes_to_visualize = len(set([item for sublist in selected_groups for item in sublist]))
    slider.config(from_=1, to=num_nodes_to_visualize, tickinterval=num_nodes_to_visualize-1)
    generate_graph(canvas, selected_groups, 1)

def add_edges_within_group(G, group):
    if isinstance(group, int):
        return
    for i, u in enumerate(group):
        for v in group[:i]:
            if isinstance(u, str) and isinstance(v, str):
                G.add_edge(u, v, weight=1)
            elif isinstance(u, tuple) and isinstance(v, tuple):
                for a in u:
                    for b in v:
                        G.add_edge(a, b, weight=1)

root = tk.Tk()
root.title("Graph Visualization")

selected_groups = []
dropdown = ttk.Combobox(root, values=options)
dropdown.grid(row=0, column=0, padx=10, pady=10)
dropdown.bind("<<ComboboxSelected>>", on_dropdown_select)

initial_num_nodes = 0
slider = tk.Scale(root, from_=0, to=initial_num_nodes, orient=tk.HORIZONTAL, length=200, tickinterval=initial_num_nodes,
                  command=lambda x: generate_graph(canvas, selected_groups, int(x)))
slider.grid(row=0, column=1, padx=10, pady=10)

fig = plt.figure(figsize=(9, 9))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().grid(row=1, column=0, columnspan=2)

generate_graph(canvas, [], 0)

root.mainloop()


```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import re

# Read the data from the CSV file
combined_csv_path = 'Encounter_combined_data.csv'
df = pd.read_csv(combined_csv_path)
df = df.head(100).replace('"', '', regex=True)

# Extract numbers from the string and create a list for each row
def extract_numbers(row):
    # Use regular expression to find all numbers
    numbers = re.findall(r'\d+', row)
    return [int(num) for num in numbers]

# Remove unnecessary characters and apply the function to create a new column 'NumberGroups'
df['NumberGroups'] = df['MergedCodes'].str.replace("'", "").str.replace(",", "").apply(extract_numbers)

# Create a graph
G = nx.Graph()

# Add edges to the graph
for group in df['NumberGroups']:
    if len(group) > 1:
        # Add edges within the group
        for i in range(len(group)-1):
            u, v = group[i], group[i+1]
            # Increase the weight of the edge if it already exists, otherwise add it with weight 1
            if G.has_edge(u, v):
                G[u][v]['weight'] += 1
            else:
                G.add_edge(u, v, weight=1)

# Draw the graph with edge thickness proportional to the weight
weights = [G[u][v]['weight'] for u, v in G.edges()]

# Adjust figure size and spacing between nodes
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, k=0.3)  # Adjust the 'k' parameter to control spacing between nodes

nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1000, edge_color='gray', width=weights, font_size=10)
plt.show()
```

```python
import tkinter as tk
from tkinter import ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
import networkx as nx
from collections import defaultdict
import math


def generate_graph(canvas, selected_groups, num_nodes_to_visualize):
    # Clear previous graph
    plt.clf()

    # Create a graph
    G = nx.Graph()

    # Add edges within each selected group
    for group in selected_groups:
        add_edges_within_group(G, group)

    # Count occurrences of each node
    node_counts = defaultdict(int)
    for group in selected_groups:
        for node in group:
            node_counts[node] += 1

    # Sort nodes based on their frequencies
    sorted_nodes = sorted(node_counts, key=node_counts.get, reverse=True)

    # Select the top num_nodes_to_visualize nodes
    selected_nodes = sorted_nodes[:num_nodes_to_visualize]

    # Extract subgraph with selected nodes and their neighbors
    subgraph = G.subgraph(selected_nodes)
    #print('subgraph', subgraph)

    # Print pairs of nodes as they occur
    print("Pairs of nodes:")
    for u, v in subgraph.edges():
        print(f"({u}, {v})")

    # Select layout algorithm based on the number of nodes
    if len(subgraph) < 10:
        pos = nx.spring_layout(subgraph)  # Use spring layout for small graphs
    else: # len(subgraph) < 50:
        pos = nx.circular_layout(subgraph)  # Use circular layout for medium-sized graphs

    #pos = nx.spring_layout(subgraph)  # Use spring layout for small graphs


    # Calculate scaling factor for node size
    #scaling_factor = 1.0 / num_nodes_to_visualize
    # Draw nodes
    node_size = [subgraph.degree(n)**2*10000/num_nodes_to_visualize**2 for n in subgraph.nodes()]  # Adjusted node size
    if num_nodes_to_visualize > 0:
        nx.draw_networkx_nodes(subgraph, pos, node_size=node_size, node_color="w", edgecolors="k", linewidths=2.0 / num_nodes_to_visualize)
    else:
        nx.draw_networkx_nodes(subgraph, pos, node_size=node_size, node_color="w", edgecolors="k", linewidths=2.0)

    # Draw node labels
    nx.draw_networkx_labels(subgraph, pos, labels={node: str(node) for node in subgraph.nodes()}, font_size=14)

    # Calculate edge weights based on some criteria
    # For example, you might calculate the weight based on the number of common neighbors between nodes u and v
    weights = [len(list(nx.common_neighbors(subgraph, u, v))) for u, v in subgraph.edges()]

    # Draw network edges with specified weights
    nx.draw_networkx_edges(subgraph, pos, width=[w**4 /num_nodes_to_visualize**2 for w in weights], edge_color="black")

    # Hide axis
    plt.axis("off")

    # Refresh canvas
    canvas.draw()

def on_dropdown_select(event):
    global selected_groups
    selected_code = dropdown.get()
    selected_groups = [group for group in groups if selected_code in group]
    print("Selected groups:", selected_groups)  # Print selected groups for debugging
    num_nodes_to_visualize = len(set([item for sublist in selected_groups for item in sublist]))
    slider.config(from_=1, to=num_nodes_to_visualize, tickinterval=num_nodes_to_visualize-1)
    generate_graph(canvas, selected_groups, 1)  # Update graph with selected groups


def add_edges_within_group(G, group):
    if isinstance(group, int):  # If only one node in the group
        return
    for i, u in enumerate(group):
        for v in group[:i]:
            if isinstance(u, str) and isinstance(v, str):  # Modify here to handle letters
                G.add_edge(u, v, weight=1)  # Add default weight of 1
            elif isinstance(u, tuple) and isinstance(v, tuple):
                for a in u:
                    for b in v:
                        G.add_edge(a, b, weight=1)  # Add default weight of 1



# Create main window
root = tk.Tk()
root.title("Graph Visualization")

#Predefined groups made up of letters including at least one of A, B, or C
# groups = [
#     ["A", "B", "C"],
#     ["B", "C", "D", "E"],
#     ["A", "B", "C", "F", "G"],
#     ["C", "D", "H"],
#     ["E", "F", "I"],
#     ["B", "C", "G", "J", "K"],
#     ["A", "D", "H", "L", "M"]
# ]

selected_groups = []

# Dropdown
#options = ["A", "B", "C"]
dropdown = ttk.Combobox(root, values=options)
dropdown.grid(row=0, column=0, padx=10, pady=10)
dropdown.bind("<<ComboboxSelected>>", on_dropdown_select)

# Slider

# Slider
initial_num_nodes = 0  # Set initial number of nodes to 0
slider = tk.Scale(root, from_=0, to=initial_num_nodes, orient=tk.HORIZONTAL, length=200, tickinterval=initial_num_nodes,
                  command=lambda x: generate_graph(canvas, selected_groups, int(x)))
slider.grid(row=0, column=1, padx=10, pady=10)

# Create a canvas to embed the matplotlib figure
fig = plt.figure(figsize=(9, 9))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().grid(row=1, column=0, columnspan=2)

# Initial visualization
generate_graph(canvas, [], 0)  # Initialize with an empty graph


# Run the Tkinter event loop
root.mainloop()
#274531002
#32485007
#736169004
```

```python
import os
import re
import pandas as pd

# Define the directory containing the FHIR files
directory = 'C:/Users/begon/synthea/output/deneme1'

# Define the regex pattern to match lines containing "resourceType": ", "id":, and "reference": "
pattern = re.compile(r'(\"resourceType\": \"(Patient|Encounter|Condition|DiagnosticReport|Observation|Procedure|Medication)\")')


# Define the list of classes to consider
classes = ['Patient', 'Encounter', 'Condition', 'DiagnosticReport', 'Observation', 'Procedure', 'Medication']

# Initialize variables to store the current resource type, ID, and references
current_resource_type = None
new_resource_type = None
current_id = None
new_id = None
current_references = []

# Initialize a dictionary to store the extracted data
data = {'ResourceType': [], 'ID': [], 'References': []}

# Initialize a dictionary to store counts for each resource type
#counts = {resource_type: 1 for resource_type in classes}

# Initialize a boolean variable to track if current_resource_type changes
INIT = False

for filename in os.listdir(directory):
    # Check if the file is a JSON file
    if filename.endswith('.json'):
        # Open the file for reading
        with open(os.path.join(directory, filename), 'r', encoding='utf-8') as f:
            # Iterate through each line in the file
            for line in f:
                # Find all matches of the pattern in the line
                match = pattern.search(line)
                # If a match is found
                if match:
                    # Check if the match belongs to one of the specified classes
                    if match.group(2) in classes:
                        # Store the resource type
                        new_resource_type = match.group(2)

                        # Check if the next line contains an ID
                        next_line = next(f, None)
                        if next_line and '"id": "' in next_line:
                            # Extract the ID from the line
                            new_id = next_line.split('"id": "')[1].split('"')[0]
#                             print('data ID',new_id)

#                             # Generate a new ID if it's a duplicate
#                             if new_id in data['ID']:
#                            current_resource_type = f"{new_resource_type[0]}"#{counts[new_resource_type]}"
                            counts[new_resource_type] += 1

                if current_resource_type == new_resource_type and '"reference": "urn:uuid:' in line:
                    # Extract the reference ID

                    # Store the reference
                    reference = line.split(':')[3].split('"')[0]  # Store the entire line
                    current_references.append(reference)       

                    
                # If a new class is encountered or it's the end of the file or resource type changes
                if current_resource_type != new_resource_type:
                    # Append the extracted data to the dictionary if current_resource_type and current_id are not None
                    if current_resource_type and current_id:
                        if current_resource_type == 'Patient':
                            data['ResourceType'].append(current_resource_type)
                        else:
                            data['ResourceType'].append(current_resource_type[0])
                        data['ID'].append(current_id)
                        data['References'].append(tuple(set(current_references)))
                        current_references = []


                    # Update current_resource_type
                    current_resource_type = new_resource_type    
                    current_id = new_id
                
                    


# Create a DataFrame from the extracted data
df = pd.DataFrame(data)


# Replace references with corresponding resource types based on IDs
for i, references_tuple in enumerate(df['References']):
    replaced_references = []
    for reference in references_tuple:
        try:
            resource_type = df.loc[df['ID'] == reference, 'ResourceType'].iloc[0]
            replaced_references.append(resource_type)
        except IndexError:
            continue
#             print(f"Reference ID '{reference}' not found in DataFrame")
#             replaced_references.append(None)  # Append None if reference ID not found
    df.at[i, 'References'] = tuple(replaced_references)




#Write the DataFrame to a CSV file
output_csv = 'output1.csv'
df.to_csv(output_csv, index=False)

df
```

```python
import networkx as nx
import matplotlib.pyplot as plt

selected_groups = [
    [1, 2, 3, 4],
    [3, 4, 5, 6]
]

# Create an empty graph
G = nx.Graph()

def add_edges_between_groups(G, selected_groups):
    # Iterate over each pair of groups
    for i, group1 in enumerate(selected_groups):
        for group2 in selected_groups[i+1:]:
            # For each pair of nodes (u, v) where u is in group1 and v is in group2
            for u in group1:
                for v in group2:
                    # Check if u and v are integers (single nodes)
                    if isinstance(u, int) and isinstance(v, int):
                        # Determine the weight of the edge based on the number of connections
                        weight = 1 if (u == v or u not in group2 or v not in group1) else 2
                        # Add an edge between u and v with the determined weight
                        G.add_edge(u, v, weight=weight)
                    # Check if u and v are tuples (multiple nodes)
                    elif isinstance(u, tuple) and isinstance(v, tuple):
                        # For each pair of nodes (a, b) where a is in u and b is in v
                        for a in u:
                            for b in v:
                                # Determine the weight of the edge based on the number of connections
                                weight = 1 if (a == b or a not in group2 or b not in group1) else 2
                                # Add an edge between a and b with the determined weight
                                G.add_edge(a, b, weight=weight)

# Add edges between selected groups
add_edges_between_groups(G, selected_groups)

# Plot the graph
pos = nx.spring_layout(G)  # Layout for positioning nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, edge_color='black', width=[2*G[u][v]['weight'] for u,v in G.edges()], linewidths=1, font_size=15)

# Display the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import math

selected_groups = [
    [1, 2, 3, 4],
    [3, 4, 5, 6]
]

# Create an empty graph
G = nx.Graph()

def add_edges_within_group(G, group):
    # Iterate over each pair of nodes in the group
    for i, u in enumerate(group):
        for v in group[i+1:]:
            # Check if the edge already exists
            if G.has_edge(u, v):
                # Increment the weight of the existing edge
                G[u][v]['weight'] += 1
            else:
                # Add a new edge with weight 1
                G.add_edge(u, v, weight=1)

# Add edges within the selected groups
for group in selected_groups:
    add_edges_within_group(G, group)

# Plot the graph
pos = nx.spring_layout(G)  # Layout for positioning nodes

# Extract edge weights to determine edge thickness
edge_weights = [G[u][v]['weight'] for u, v in G.edges()]

# Calculate exponential values of edge weights
exp_edge_weights = [math.exp(weight) for weight in edge_weights]

# Print edge weights and exponential values
print("edge_weights:", edge_weights)
print("exp_edge_weights:", exp_edge_weights)

# Calculate adjusted linewidths
linewidths = [weight * 2 for weight in exp_edge_weights]
print("linewidths:", linewidths)

# Draw nodes and edges with variable thickness based on exponential of edge weights
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, 
        edge_color='black', width=linewidths, linewidths=linewidths, font_size=15)

# Display the plot
plt.show()
```
